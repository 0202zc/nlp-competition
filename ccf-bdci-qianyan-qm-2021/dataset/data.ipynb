```python
train = pd.read_csv('./data/train.tsv', sep='\t', encoding='utf8')
dev = pd.read_csv('./data/dev.tsv', sep='\t', encoding='utf8')
train, dev
```

```python
sns.countplot('label', data=train)
plt.title('bq + lcqmc + oppo training set')
# plt.savefig('15-21kind.png', dpi=150)
plt.show()
```

```python
sns.countplot('label', data=dev)
plt.title('bq + lcqmc + oppo val set')
# plt.savefig('15-21kind.png', dpi=150)
plt.show()
```

```python
train_length = []
dev_length = []
paws_x_train_length = []
paws_x_dev_length = []

for i in range(train.shape[0]):
    train_length.append(len(str(train['q1'].iloc[i])) + len(str(train['q2'].iloc[i])))

for i in range(dev.shape[0]):
    dev_length.append(len(str(dev['q1'].iloc[i])) + len(str(dev['q2'].iloc[i])))

for i in range(paws_x_train.shape[0]):
    paws_x_train_length.append(len(str(paws_x_train['q1'].iloc[i])) + len(str(paws_x_train['q2'].iloc[i])))

for i in range(paws_x_dev.shape[0]):
    paws_x_dev_length.append(len(str(paws_x_dev['q1'].iloc[i])) + len(str(paws_x_dev['q2'].iloc[i])))
```

```python
plt.style.use('fivethirtyeight')
sns.distplot(train_length)
```

```python
plt.style.use('fivethirtyeight')
sns.distplot(dev_length)
```

```python
plt.style.use('fivethirtyeight')
sns.distplot(paws_x_train_length)
```

```python
plt.style.use('fivethirtyeight')
sns.distplot(paws_x_dev_length)
```
