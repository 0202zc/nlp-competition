```python
import json
```

```python
json_data = []
for line in open(file='train.json', mode='r', encoding='utf8'):
    json_data.append(json.loads(line))


convert_data = []

for data in json_data:
    convert_data.append(data['sentence1'] + '\t' + data['sentence2'] + '\t' + data['label'])

with open('train.csv', 'w', encoding='utf8') as f:
    for row in convert_data:
        f.write(row + '\n')

json_data, convert_data
```
