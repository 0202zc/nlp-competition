#### 测试集按照训练集的10%采样

假标签(Pseudo Label)
假标签是一种半监督方法：

通过训练集训练好的模型，去预测测试集的结果
把测试集这个结果作为测试集的标签，和训练集混合在一起训练模型
一般模型能忍受10%的噪音，所以建议假标签方法不要一股脑把所有测试集和训练集混合，尽量保持比例在10：1附近。

```python
train_df = pd.read_csv('translation/zh-CN/generate/train_transback_split.tsv', encoding='utf8', sep='\t')
testA_df = pd.read_csv('test/split/testA_6453.tsv', sep='\t', encoding='utf8', header=0, names=['label_id', 'text'])
for i in range(36):
    # testA_label_i = pd.read_csv('labels/testA/split/%d.tsv' % i, sep='\t', encoding='utf8')
    train_i = train_df[train_df['label_id'] == i]
    testA_i = testA_df[testA_df['label_id'] == i]
    num_i = train_i.shape[0], testA_i.shape[0]

    if num_i[0] * 0.1 <= num_i[1]:
        train_df = pd.concat((train_df, testA_i.head(int(num_i[0] * 0.1))), axis=0)
        print(int(num_i[0] * 0.1), end='\t')

train_df.sort_values('label_id').to_csv("train/split/train_backtrans_6453(0.1).tsv", sep='\t', encoding='utf8', index=False)
```

#### 按比例采样

```python
import numpy as np
import pandas as pd

# train_df = pd.read_csv('train_xlnet.tsv', sep='\t', encoding='utf8')
train_df = pd.read_json('train.json', encoding='utf8', lines=True)  # 参考比例
# train_df = pd.read_csv('translation/zh-CN/generate/train_transback_split.tsv', sep='\t', encoding='utf8')
train_labeled_df = pd.read_csv('train/split/B/field/3/trainB3_backtrans_field_abstract(simplify).tsv', sep='\t', encoding='utf8')

ratio = 49.0

temp_train_df = pd.DataFrame()

for i in range(36):
    # 获取当前比例
    new, origin = train_labeled_df[train_labeled_df['label_id'] == i].shape[0], train_df[train_df['label_id'] == i].shape[0]
    current_ratio = new / origin
    # 如果不够，则补充
    if current_ratio < ratio - 4.0:
        add_data = train_labeled_df[train_labeled_df['label_id'] == i]
        temp_train_df = pd.concat((temp_train_df, add_data), axis=0)
        need = origin * ratio - new

        label = train_df[train_df['label_id'] == i]
        # 如果 label 资源整体数目小于需求
        if origin < need:
            times, rest = int(need // origin), int(need % origin)
            # 添加 times 倍的全量数据
            for t in range(times):
                add_data = train_labeled_df[train_labeled_df['label_id'] == i].iloc[:origin, :]
                temp_train_df = pd.concat((temp_train_df, add_data), axis=0)
            # 添加 rest 的余量数据
            add_data = train_labeled_df[train_labeled_df['label_id'] == i].iloc[:rest, :]
            temp_train_df = pd.concat((temp_train_df, add_data), axis=0)
        # 如果满足需求，直接添加
        else:
            add_data = train_labeled_df[train_labeled_df['label_id'] == i].iloc[:need, :]
            temp_train_df = pd.concat((temp_train_df, add_data), axis=0)
    else:
        # 如果超过一定的比例，则削减
        ALLOW = 50.0
        if current_ratio - ratio > ALLOW:
            need = int(origin * (ratio + ALLOW))
            add_data = train_labeled_df[train_labeled_df['label_id'] == i].iloc[:need, :]
            temp_train_df = pd.concat((temp_train_df, add_data), axis=0)
        else:
            add_data = train_labeled_df[train_labeled_df['label_id'] == i]
            temp_train_df = pd.concat((temp_train_df, add_data), axis=0)

ratios = []
for i in range(36):
    ratio = temp_train_df[temp_train_df['label_id'] == i].shape[0] / train_df[train_df['label_id'] == i].shape[0]
    ratios.append(ratio)
    print("%d\t%f" % (i, ratio))

temp_train_df.to_csv('train/split/B/field/3/trainB3_backtrans_field_sampling49.tsv', sep='\t', encoding='utf8', index=None)
np.mean(ratios)
```

```python
import numpy as np

def get_sep_abstracts(abstracts, token_sep='<sep>'):
    abstract_results = []
    # 分割 effect
    for abstract in abstracts:
        result1 = abstract.rsplit(sep='；', maxsplit=1)
        result2 = abstract.rsplit(sep='。', maxsplit=2)
        result3 = abstract.rsplit(sep='，', maxsplit=2)

        if len(result2) < 3:
            if len(result1) == 2:
                # 再按照分号分割
                abstract_results.append(result1[0] + token_sep + result1[1])
                # print(result1[0] + token_sep + result1[1])
            else:
                # 最后按照逗号分割
                abstract_results.append(result3[0] + token_sep + result3[1] if len(result3) > 1 else result3[0])
                # print(result3[0] + token_sep + result3[1] if len(result3) > 1 else result3[0])
        else:
            # 先按照句号分割
            abstract_results.append(result2[0] + token_sep + result2[1])
            # print(result2[0] + token_sep + result2[1])

    abstract_results_2, abstract_results_3 = [], []
    # 分割 title 和 program
    for abstract in abstract_results:
        symbols = [token_sep, '。', '，', '；']
        pos0, pos1, pos2, pos3 = [abstract.find(symbol) for symbol in symbols]
        positions = [pos0, pos1, pos2, pos3]
        if pos0 == -1:
            abstract_results_2.append(abstract)
            continue

        # 如果title已被上面程序分割，则跳过
        # if pos1 == -1 and pos2 == -1 or pos0 < pos1 and pos0 < pos2 and pos0 < pos3:
        #     continue

        result = []

        if pos1 != -1:
            if pos2 != -1:
                if pos3 != -1:
                    result = abstract.split(symbols[np.argmin(np.asarray(positions))], maxsplit=1)
                else:
                    result = abstract.split(symbols[np.argmin(np.asarray(positions.pop(symbols.index('；'))))],
                                            maxsplit=1)
            else:
                # positions.pop(symbols.index('，'))
                positions[symbols.index('，')] = 8888
                if pos3 != -1:
                    result = abstract.split(symbols[np.argmin(np.asarray(positions))], maxsplit=1)
                else:
                    result = abstract.split(symbols[np.argmin(np.asarray(positions.pop(symbols.index('；'))))],
                                            maxsplit=1)
        else:
            # positions.pop(symbols.index('。'))
            positions[symbols.index('。')] = 8888
            if pos2 != -1:
                if pos3 != -1:
                    result = abstract.split(symbols[np.argmin(np.asarray(positions))], maxsplit=1)
                else:
                    result = abstract.split(symbols[np.argmin(np.asarray(positions.pop(symbols.index('；'))))],
                                            maxsplit=1)
            else:
                # positions.pop(symbols.index('，'))
                positions[symbols.index('，')] = 8888
                if pos3 != -1:
                    result = abstract.split(symbols[np.argmin(np.asarray(positions))], maxsplit=1)
                else:
                    result = abstract.split(symbols[np.argmin(np.asarray(positions.pop(symbols.index('；'))))],
                                            maxsplit=1)

        abstract_results_2.append(result[0] + token_sep + result[1] if result != [] else abstract)

    for abstract in abstract_results_2:
        symbols = [token_sep, '。', '，', '；']
        pos0, pos1, pos2, pos3 = [abstract.find(symbol) for symbol in symbols]
        if pos0 == -1 or pos0 < pos2:
            abstract_results_3.append(abstract)
            continue

        result = []

        result = abstract.split('，', maxsplit=1)
        abstract_results_3.append(result[0] + token_sep + result[1] if len(result) > 1 else abstract)

    return abstract_results_3


def totally_sep(abstracts: list, token_sep='<sep>'):
    abstract_results = []

    for abstract in abstracts:
        abstract = abstract.replace('。', token_sep).replace('；', token_sep)[:-len(token_sep)] + '。'
        # 分割每个句子的第一个和最后N个逗号
        # separate = abstract.split('，', maxsplit=1)
        # abstract = separate[0] + token_sep + separate[1] if len(separate) > 1 else abstract
        # separate = abstract.rsplit('，', maxsplit=1)
        # abstract = separate[0] + token_sep + separate[1] if len(separate) > 1 else abstract
        # abstract = token_sep.join(abstract.split('，', maxsplit=1))
        # abstract = token_sep.join(abstract.rsplit('，', maxsplit=2))
        abstract_results.append(abstract)

    return abstract_results
```

```python
import pandas as pd

def correlation_calculating(file1: str, file2: str) -> list:
    result = []
    if file1.__contains__('txt'):
        result1 = pd.read_csv(file1, header=None, names=['label'], encoding='utf8')
    else:
        result1 = pd.read_csv(file1, encoding='utf8')
    if file2.__contains__('txt'):
        result2 = pd.read_csv(file2, header=None, names=['label'], encoding='utf8')
    else:
        result2 = pd.read_csv(file2, encoding='utf8')
    same, sum = result1[result1['label'] == result2['label']].shape[0], result1.shape[0]
    # result = result2[result1['label'] != result2['label']]
    result = pd.concat((result1[result1['label'] != result2['label']], result2[result1['label'] != result2['label']]), axis=1)
    correlation = same / sum

    # return [result, correlation, same, sum, sum - same]
    return result


res = correlation_calculating("../results/B/3(ema0.952)/submit (test3_ema0.952 manually).csv", "../results/B/5(sampling49+50+2)/test_results_0.txt")
pd.DataFrame(res.head(-1)).reset_index(drop=True)
```
