```python
import pandas as pd
import numpy as np

df = pd.read_json('发明专利数据.json', encoding='utf8', lines=False).fillna("")
df
```

```python
token_sep = '<sep>'
df['text'] = df['pat_name'].str.replace('<i>|</i>|\n|\t|<br>', '', regex=True).str.strip() + token_sep + df['pat_summary'].str.replace('<i>|</i>|\n|\t|<br>', '', regex=True)
df['text'].head(40000).to_csv('2017patents_40000.tsv', sep='\t', encoding='utf8', index=False)
df['text']
```

```python
text = np.asarray(df['text'].head(40000)).transpose().tolist()
text_length = [len(a) for a in text]
text_length.sort(reverse=True)

import seaborn as sns
from matplotlib import pyplot as plt

plt.style.use('fivethirtyeight')
plt.figure(figsize=(20, 15))
sns.set(font='SimHei', font_scale=1.5)  # 解决Seaborn中文显示问题并调整字体大小
# sns.distplot(abstract_length)
sns.displot(data=text_length, kind='hist', kde=True), max(text_length), min(text_length), np.median(text_length), np.mean(text_length), text_length
```

```python
labels = pd.read_csv('40000/label_40000.txt', sep='\t', encoding='utf8', header=None, names=['label'])
train_df = pd.read_csv('40000/2017patents_40000.tsv', sep='\t', encoding='utf8')
train_df['label_id'] = labels['label']
train_df[['label_id', 'text']].to_csv('40000/2017patents_40000_labeled.tsv', sep='\t', encoding='utf8', index=False)
train_df
```
