```python
ratio = 24.0

temp_train_df = pd.DataFrame()

for i in range(36):
    # 获取当前比例
    new, origin = train_labeled_df[train_labeled_df['label_id'] == i].shape[0], train_df[train_df['label_id'] == i].shape[0]
    current_ratio = new / origin
    # 如果不够，则补充
    if current_ratio < ratio - 3.0:
        add_data = train_labeled_df[train_labeled_df['label_id'] == i]
        temp_train_df = pd.concat((temp_train_df, add_data), axis=0)
        need = origin * ratio - new

        label = train_df[train_df['label_id'] == i]
        # 如果 label 资源整体数目小于需求
        if origin < need:
            times, rest = int(need // origin), int(need % origin)
            # 添加 times 倍的全量数据
            for t in range(times):
                add_data = train_labeled_df[train_labeled_df['label_id'] == i].iloc[:origin, :]
                temp_train_df = pd.concat((temp_train_df, add_data), axis=0)
            # 添加 rest 的余量数据
            add_data = train_labeled_df[train_labeled_df['label_id'] == i].iloc[:rest, :]
            temp_train_df = pd.concat((temp_train_df, add_data), axis=0)
        # 如果满足需求，直接添加
        else:
            add_data = train_labeled_df[train_labeled_df['label_id'] == i].iloc[:need, :]
            temp_train_df = pd.concat((temp_train_df, add_data), axis=0)
    else:
        # 如果超过一定的比例，则削减
        ALLOW = 16.0
        if current_ratio - ratio > ALLOW:
            need = int(origin * (ratio + ALLOW))
            add_data = train_labeled_df[train_labeled_df['label_id'] == i].iloc[:need, :]
            temp_train_df = pd.concat((temp_train_df, add_data), axis=0)
        else:
            add_data = train_labeled_df[train_labeled_df['label_id'] == i]
            temp_train_df = pd.concat((temp_train_df, add_data), axis=0)

ratios = []
for i in range(36):
    ratio = temp_train_df[temp_train_df['label_id'] == i].shape[0] / train_df[train_df['label_id'] == i].shape[0]
    ratios.append(ratio)
    print("%d\t%f" % (i, ratio))

temp_train_df.to_csv('train/split/train6451_backtras_oversampling24.tsv', sep='\t', encoding='utf8', index=None)
np.mean(ratios)
```

```python
import numpy as np
import pandas as pd

train_df = pd.read_json('train.json', encoding='utf8', lines=True)
testA_df = pd.read_json('testA.json', encoding='utf8', lines=True)

abstracts_train = train_df['abstract'].tolist()
abstracts_testA = testA_df['abstract'].tolist()

token_sep = '<sep>'


def get_sep_abstracts(abstracts, token_sep='<sep>'):
    abstract_results = []
    # 分割 effect
    for abstract in abstracts:
        result1 = abstract.rsplit(sep='；', maxsplit=1)
        result2 = abstract.rsplit(sep='。', maxsplit=2)
        result3 = abstract.rsplit(sep='，', maxsplit=2)

        if len(result2) < 3:
            if len(result1) == 2:
                # 再按照分号分割
                abstract_results.append(result1[0] + token_sep + result1[1])
                # print(result1[0] + token_sep + result1[1])
            else:
                # 最后按照逗号分割
                abstract_results.append(result3[0] + token_sep + result3[1] if len(result3) > 1 else result3[0])
                # print(result3[0] + token_sep + result3[1] if len(result3) > 1 else result3[0])
        else:
            # 先按照句号分割
            abstract_results.append(result2[0] + token_sep + result2[1])
            # print(result2[0] + token_sep + result2[1])

    abstract_results_2, abstract_results_3 = [], []
    # 分割 title 和 program
    for abstract in abstract_results:
        symbols = [token_sep, '。', '，', '；']
        pos0, pos1, pos2, pos3 = [abstract.find(symbol) for symbol in symbols]
        positions = [pos0, pos1, pos2, pos3]
        if pos0 == -1:
            abstract_results_2.append(abstract)
            continue

        # 如果title已被上面程序分割，则跳过
        # if pos1 == -1 and pos2 == -1 or pos0 < pos1 and pos0 < pos2 and pos0 < pos3:
        #     continue

        result = []

        if pos1 != -1:
            if pos2 != -1:
                if pos3 != -1:
                    result = abstract.split(symbols[np.argmin(np.asarray(positions))], maxsplit=1)
                else:
                    result = abstract.split(symbols[np.argmin(np.asarray(positions.pop(symbols.index('；'))))],
                                            maxsplit=1)
            else:
                # positions.pop(symbols.index('，'))
                positions[symbols.index('，')] = 8888
                if pos3 != -1:
                    result = abstract.split(symbols[np.argmin(np.asarray(positions))], maxsplit=1)
                else:
                    result = abstract.split(symbols[np.argmin(np.asarray(positions.pop(symbols.index('；'))))],
                                            maxsplit=1)
        else:
            # positions.pop(symbols.index('。'))
            positions[symbols.index('。')] = 8888
            if pos2 != -1:
                if pos3 != -1:
                    result = abstract.split(symbols[np.argmin(np.asarray(positions))], maxsplit=1)
                else:
                    result = abstract.split(symbols[np.argmin(np.asarray(positions.pop(symbols.index('；'))))],
                                            maxsplit=1)
            else:
                # positions.pop(symbols.index('，'))
                positions[symbols.index('，')] = 8888
                if pos3 != -1:
                    result = abstract.split(symbols[np.argmin(np.asarray(positions))], maxsplit=1)
                else:
                    result = abstract.split(symbols[np.argmin(np.asarray(positions.pop(symbols.index('；'))))],
                                            maxsplit=1)

        abstract_results_2.append(result[0] + token_sep + result[1] if result != [] else abstract)

    for abstract in abstract_results_2:
        symbols = [token_sep, '。', '，', '；']
        pos0, pos1, pos2, pos3 = [abstract.find(symbol) for symbol in symbols]
        if pos0 == -1 or pos0 < pos2:
            abstract_results_3.append(abstract)
            continue

        result = []

        result = abstract.split('，', maxsplit=1)
        abstract_results_3.append(result[0] + token_sep + result[1] if len(result) > 1 else abstract)

    return abstract_results_3


def totally_sep(abstracts: list, token_sep='<sep>'):
    abstract_results = []

    for abstract in abstracts:
        abstract = abstract.replace('。', token_sep).replace('；', token_sep)[:-len(token_sep)] + '。'
        # 分割每个句子的第一个和最后N个逗号
        # separate = abstract.split('，', maxsplit=1)
        # abstract = separate[0] + token_sep + separate[1] if len(separate) > 1 else abstract
        # separate = abstract.rsplit('，', maxsplit=1)
        # abstract = separate[0] + token_sep + separate[1] if len(separate) > 1 else abstract
        # abstract = token_sep.join(abstract.split('，', maxsplit=1))
        # abstract = token_sep.join(abstract.rsplit('，', maxsplit=2))
        abstract_results.append(abstract)

    return abstract_results


if __name__ == '__main__':
    assignee = False
    totally = False

    if totally:
        abstracts_train_df = pd.DataFrame(data=totally_sep(abstracts_train), columns=['abstract'])
        train_df['abstract'] = abstracts_train_df['abstract']
        train_df['title'] = train_df['title'].apply(lambda x: x.strip())
        train_df['abstract'] = train_df['abstract'].fillna('').apply(lambda x: x.strip())
        if assignee:
            train_df['assignee'] = train_df['assignee'].apply(lambda x: x.strip())
            train_df['text'] = train_df['title'].str.lower() + token_sep + train_df[
                'assignee'].str.lower() + token_sep + train_df['abstract'].str.lower()
        else:
            train_df['text'] = train_df['title'].str.lower() + token_sep + train_df['abstract'].str.lower()
        train_df['text'] = train_df['text'].str.replace('<i>|</i>', '', regex=True)

        train_df[['label_id', 'text']].to_csv('split/totally/train_split.tsv', index=None, sep='\t')

        abstracts_testA_df = pd.DataFrame(data=totally_sep(abstracts_testA), columns=['abstract'])
        testA_df['abstract'] = abstracts_testA_df['abstract']
        testA_df['title'] = testA_df['title'].apply(lambda x: x.strip())
        testA_df['abstract'] = testA_df['abstract'].fillna('').apply(lambda x: x.strip())
        if assignee:
            testA_df['assignee'] = testA_df['assignee'].fillna('').apply(lambda x: x.strip())
            testA_df['text'] = testA_df['title'].str.lower() + token_sep + testA_df[
                'assignee'].str.lower() + token_sep + testA_df['abstract'].str.lower()
        else:
            testA_df['text'] = testA_df['title'].str.lower() + token_sep + testA_df['abstract'].str.lower()
        testA_df['text'] = testA_df['text'].str.replace('<i>|</i>', '', regex=True)

        testA_df[['text']].to_csv('split/totally/testA_split.tsv', index=None, sep='\t')

        label = pd.read_csv(
            "../results/60/4/submit (6448_bactrans_oversampling24_4epochs_rdrop result_2 6451).csv",
            encoding='utf8')
        testA_df['label'] = label['label']
        testA_df['title'] = testA_df['title'].apply(lambda x: x.strip())
        testA_df['abstract'] = testA_df['abstract'].fillna('').apply(lambda x: x.strip())
        if assignee:
            testA_df['assignee'] = testA_df['assignee'].fillna('').apply(lambda x: x.strip())
            testA_df['text'] = testA_df['title'].str.lower() + token_sep + testA_df[
                'assignee'].str.lower() + token_sep + testA_df['abstract'].str.lower()
        else:
            testA_df['text'] = testA_df['title'].str.lower() + token_sep + testA_df['abstract'].str.lower()
        testA_df['text'] = testA_df['text'].str.replace('<i>|</i>', '', regex=True)
        testA_df[['label', 'text']].to_csv('test/split/totally/testA_6451.tsv', index=None, sep='\t')
    else:
        abstracts_train_df = pd.DataFrame(data=get_sep_abstracts(abstracts_train), columns=['abstract'])
        train_df['abstract'] = abstracts_train_df['abstract']
        train_df['title'] = train_df['title'].apply(lambda x: x.strip())
        train_df['abstract'] = train_df['abstract'].fillna('').apply(lambda x: x.strip())
        if assignee:
            train_df['assignee'] = train_df['assignee'].apply(lambda x: x.strip())
            train_df['text'] = train_df['title'].str.lower() + token_sep + train_df[
                'assignee'].str.lower() + token_sep + train_df['abstract'].str.lower()
        else:
            train_df['text'] = train_df['title'].str.lower() + token_sep + train_df['abstract'].str.lower()
        train_df['text'] = train_df['text'].str.replace('<i>|</i>', '', regex=True)

        train_df[['label_id', 'text']].to_csv('split/train_split.tsv', index=None, sep='\t')

        abstracts_testA_df = pd.DataFrame(data=get_sep_abstracts(abstracts_testA), columns=['abstract'])
        testA_df['abstract'] = abstracts_testA_df['abstract']
        testA_df['title'] = testA_df['title'].apply(lambda x: x.strip())
        testA_df['abstract'] = testA_df['abstract'].fillna('').apply(lambda x: x.strip())
        if assignee:
            testA_df['assignee'] = testA_df['assignee'].fillna('').apply(lambda x: x.strip())
            testA_df['text'] = testA_df['title'].str.lower() + token_sep + testA_df[
                'assignee'].str.lower() + token_sep + testA_df['abstract'].str.lower()
        else:
            testA_df['text'] = testA_df['title'].str.lower() + token_sep + testA_df['abstract'].str.lower()
        testA_df['text'] = testA_df['text'].str.replace('<i>|</i>', '', regex=True)

        testA_df[['text']].to_csv('split/testA_split.tsv', index=None, sep='\t')

        label = pd.read_csv(
            "../results/60/4/submit (6448_bactrans_oversampling24_4epochs_rdrop result_2 6451).csv",
            encoding='utf8')
        testA_df['label'] = label['label']
        testA_df['title'] = testA_df['title'].apply(lambda x: x.strip())
        testA_df['abstract'] = testA_df['abstract'].fillna('').apply(lambda x: x.strip())
        if assignee:
            testA_df['assignee'] = testA_df['assignee'].fillna('').apply(lambda x: x.strip())
            testA_df['text'] = testA_df['title'].str.lower() + token_sep + testA_df[
                'assignee'].str.lower() + token_sep + testA_df['abstract'].str.lower()
        else:
            testA_df['text'] = testA_df['title'].str.lower() + token_sep + testA_df['abstract'].str.lower()
        testA_df['text'] = testA_df['text'].str.replace('<i>|</i>', '', regex=True)
        testA_df[['label', 'text']].to_csv('test/split/testA_6451.tsv', index=None, sep='\t')

abstracts_train_df['abstract']
```

```python
import time
import json
from tencentcloud.common.exception.tencent_cloud_sdk_exception import TencentCloudSDKException
from tencentcloud.common.abstract_client import AbstractClient
from tencentcloud.tmt.v20180321 import models, tmt_client
from tencentcloud.common import credential
from tencentcloud.common.profile.client_profile import ClientProfile
from tencentcloud.common.profile.http_profile import HttpProfile


def tencent_translate(content, src='zh', tgt="en"):
    """腾讯翻译引擎, 详细API接口：https://cloud.tencent.com/document/api/551/15619"""
    time.sleep(0.1)
    try:
        cred = credential.Credential("*********************************",
                                     "*********************************")  #自己申请的api和密钥
        httpProfile = HttpProfile()
        httpProfile.endpoint = "tmt.tencentcloudapi.com"
        clientProfile = ClientProfile()
        clientProfile.httpProfile = httpProfile
        client = tmt_client.TmtClient(cred, "ap-shanghai", clientProfile)
        # if len(content) >= 200:
        #     content = content[:200]
        req = models.TextTranslateRequest()
        params = {
            "SourceText": content,
            "Source": src,
            "Target": tgt,
            "ProjectId": 0
        }
        req.from_json_string(json.dumps(params))
        resp = client.TextTranslate(req)
        result = resp.to_json_string()
        result = eval(result)
        return result["TargetText"]
    except TencentCloudSDKException as err:
        print(err.message)
        return None
```
